```python
def comma(df):
    if pd.isna(df):
        return np.nan
    elif df.strip()=='':
        return np.nan
    elif df.find(',')!=-1:
        return int(df.replace(',',''))
    else:
        return int(df)
```

```python
comma(df_last.loc[0, '분양가격(제곱미터)'])
comma(df_last.loc[9093, '분양가격(제곱미터)'])
```

```python
df_last['제곱미터당가격'] = df_last['분양가격(제곱미터)'].map(comma)
df_last.head()
```

```python
df_last['평당분양가격'] = df_last['제곱미터당가격']*3.3
df_last
```
