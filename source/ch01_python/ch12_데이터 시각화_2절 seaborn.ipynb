## 2.3 Seaborn 함수로 그래프 그리기
### 1) 관계형 그래프
    - x축이 수치형 데이터
   (ex) scatterplot, lineplot => relplot(서브플롯)

````python
sns.scatterplot(data=iris, x='petal_length', 
                y='petal_width', 
                hue='species', 
                palette='Set1', 
                style='species')
# species 별 색생과 스타일이 다른 scatterplot
plt.show()
````

````python
# 종별 lineplot 
# lineplot 의 음영 : 95% 신뢰구간
sns.lineplot(data=iris, x='petal_length', 
             y='petal_width', 
             hue='species', 
             palette='tab10', 
             style='species')
plt.show()
````

````python
# 종별 lineplot을 서브플롯으로
sns.relplot(data=iris, x='petal_length', 
            y='petal_width', 
            hue='species',
            # kind='line', # kind='scatter가 기본값
            col='species',   # species별 서브플롯
           #col_wrap=2) # 한줄에 표현될 서브플롯 수
           )
plt.show()
````

````python
# line과 scatter를 종별 서브플롯으로 겹쳐서
g = sns.relplot(data=iris, x='petal_length', y='petal_width', 
            hue='species',
            # kind='line', # kind='scatter가 기본값
            col='species',   # species별 서브플롯
           #col_wrap=2) # 한줄에 표현될 서브플롯 수
               )
# g그래프에 line플롯을 추가
g.map_dataframe(sns.lineplot, x='petal_length', y='petal_width')
# 크기 조정 : 서브플롯함수에서는 내부적으로 figure 새로 만들기 때문에 figsize 무시
g.fig.set_size_inches(10,3)              
````

````python
sns.scatterplot(data=iris, x='petal_length', 
                y='petal_width', 
                hue='species', 
                palette='Set1', 
                style='species')
# species 별 색생과 스타일이 다른 scatterplot
sns.lineplot(data=iris, x='petal_length', 
             y='petal_width', 
             hue='species', 
             palette='tab10', 
             style='species')
plt.show()
````

````python
# ax[0]에는 scatter, ax[1] lineplot
fig, ax = plt.subplots(1, ncols=2)
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=ax[0])
sns.lineplot(data=iris, x='petal_length', y='petal_width', hue='species', ax=ax[1])
plt.show()
````

### 2) 범주형 그래프
    - x축이 범주형 데이터
```
   (ex) stripplot(무작위 흩뿌림), swarmplot(겹치지않는 선점도), barplot, barhplot(y의 평균 막대 그래프), 
        pointplot(y의 평균으로 line), countplot(count는 x의 빈도수가 높이), 
        boxplot, violonplot, aoxenplot
        => catplot(서브플롯)
```

````python
# x축이 범주형일 때의 산점도
sns.stripplot(data=iris, x='species', y='petal_width', hue='species')
plt.show()
````

````python
# x축이 범주형일 때의 산점도(분포 포함)
sns.swarmplot(data=iris, x='species', y='petal_width', hue='species')
plt.show()
````

````python
# 범주형 그래프의 x축을 지정하지 않으면
sns.swarmplot(data=iris, x='species', y='petal_width', hue='species', order=['versicolor','setosa','virginica'])
plt.show()
````

````python
# barplot, barhplot, pointplot : 각 범주별 평균값(y)과 그애 대헌 신뢰구간
            # 신뢰구간 제외 : errorbar=none

sns.barplot(data=iris, x='species', y='petal_width', #errorbar=None,
           )
sns.lineplot(data=iris, x='species', y='petal_width', color='k'
            )
plt.show()
````

````python
sns.pointplot(data=iris, x='species', y='petal_width')  # 신뢰구간을 제외
````

## 2.4 다중그래프를 위한 FacetGrid

````python
g = sns.FacetGrid(iris, col='species', hue='species')
g.map(sns.scatterplot, 'petal_length','petal_width')
plt.show()
````
